## Synchronising Image Data Formatting

```python
# Picture data with values already divided by 255 are rescaled for uniformity, and cast back into integer
df_comb['image_data'] = [(df_comb['image_data'][x]*255).astype(int) if df_comb['image_data'][x].dtype == 'float32' else (df_comb['image_data'][x]) for x in range(len(df_comb['image_data']))] 
```

## Padding images to same size

```python
# function for padding images to same size of max_allowable_pixels_width = 1024, max_allowable_pixels_height = 1024

max_allowable_pixels_height = 256
max_allowable_pixels_width = 256
color_channels = 3


def padding_func(image_data): # array of (height pixels by width pixels by color channels)

    # read image
    img = image_data
    ht, wd, cc= img.shape[0], img.shape[1], img.shape[2]

    # create new image of desired size and color (black) for padding
    ww = max_allowable_pixels_width
    hh = max_allowable_pixels_height
    color = (0,0,0)
    padded_image_data = np.full((hh,ww,cc), color, dtype=np.uint8)

    # compute center offset
    xx = (ww - wd) // 2
    yy = (hh - ht) // 2

    # copy img image into center of result image
    padded_image_data[yy:yy+ht, xx:xx+wd] = img

    # sending back result into DataFrame
    return padded_image_data
```

```python
len(df_comb['image_size'][1])
```

```python
## dropping all images with image_size not having 3 components
df_comb = df_comb[df_comb['image_size'].map(len) == 3]
```

```python
# resetting index after dropping non 3-component image_size rows
df_comb.reset_index(drop=True, inplace=True)
```

```python
## dropping all images with any dimension pixel count more than max_allowable_pixels_height, max_allowable_pixels_width and color_channels not 3

df_comb = df_comb[df_comb.apply(lambda row: row['image_size'][0] <= max_allowable_pixels_height and row['image_size'][1] <= max_allowable_pixels_width and row['image_size'][2] == color_channels, axis=1)]
```

```python
# resetting index after dropping images larger than more than max_allowable_pixels_height, max_allowable_pixels_width and color_channels not 3

df_comb.reset_index(drop=True, inplace=True)
```

```python
# testing function with one row of image data
sample_image = df_comb['image_data'][0]

# show pre-padded image
plt.figure(figsize = (15,5))
plt.imshow(sample_image)
plt.show();

# show padded image
plt.figure(figsize = (30,10))
plt.imshow(padding_func(sample_image))
plt.show();
```

```python
df_comb['padded_image_data'] = [padding_func(df_comb['image_data'][x]) for x in range(len(df_comb['image_data']))]
```

## Scaling and Train Test Split

```python
X = pd.DataFrame(df_comb['padded_image_data'])
X.head()
```

```python
y = pd.DataFrame(df_comb['modified'])
```

```python
%%time
# The current range of X is 0 to 255.
# The code below is equivalent to X_train = X_train / 255.
# This scales each value to be between 0 and 1.
X /= 255.
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

```python
## resetting indexes for X_train, X_test, y_train, y_test

X_train.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
y_train.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)
```
